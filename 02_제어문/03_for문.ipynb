```python
    # for 문의 기본 구조
# for 변수 in 리스트(또는 튜플, 문자열):
#     수행할 문장 1
#     수행할 문장 2
```

```python
test_list = ["one", "two", "three"]

for tl in test_list:
    print(tl)
```

```python
a = [(1, 2), (3, 4), (5, 6)]
for (j, k) in a:
    print(j)
```

```python
marks = [90, 25, 67, 45, 80]

number = 0
for mark in marks :
    number += 1
    if mark > 60:
        print(f"{number}번 학생은 합격입니다.")
    else:
        print(f"{number}번 학생은 불합격입니다.")
```

```python
    # for 문과 continue

marks = [90, 25, 67, 45, 80]
number = 0
for mark in marks :
    number += 1
    if mark <60 :
        continue
    print(f"{number}번 학생은 합격입니다.")
```

```python
    # for 문과 range 함수

b = range(1, 10+1, 2)
list(b)
c = range(10, 0, -2)
list(c)
```

```python
    # for 문과 range를 이용한 구구단
for i in range(7, 9+1):
    for j in range(1, 9+1):
        print(f"{i} X {j} =",i * j)
```

```python
    # 리스트 컴프리헨션 사용하기
a = [1, 2, 3, 4]
result = []

for num in a:
    result.append(num * 3)

result
```

```python
a = [1, 2, 3, 4]
result = [num * 3 for num in a]
print(result)
```

```python
    # for 문과 break 문
for i in range(10):
    if i == 5:
        break
    else:
        print(i)
```

```python
    # enumerate() 함수 -> 인덱스 번호를 줌
fruits = ['apple', 'banana', 'orange']
for (i, fruit) in enumerate(fruits):
    print(f"{i}번째 과일 : {fruit}")
```

```python
    # zip 함수로 여러 리스트 순회하기
names = ['홍길동', '김철수', '이영희']
scores = [85, 92, 78]

for name, score in zip(names, scores):
    print(f"{name} : {score}점")
```
