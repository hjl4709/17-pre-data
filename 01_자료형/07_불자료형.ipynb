```python
    # 불(bool) 자료형은 어떻게 사용할까?
a = True # T & F는 대문자!
b = False
print(a, b)
print(type(a), type(b))
```

```python
print(1 == 1)
print(2 > 1)
print(2 < 1)
print(2 >= 2)
print(2 <= 2)
print(2 != 2)
```

```python
    # if 조건문: 
    #     수행할 문장 

print(bool("python"))
print(bool("")) # 공백 -> 문자가 없기 때문에 False
print(bool(" "))

print(bool(10))
print(bool(0)) # 0 -> 숫자는 0만 False

print(bool([1, 2, 3]))
print(bool([])) # 비어있는 리스트만 False

print(bool({"pey"}))
print(bool({})) # 비어있는 딕셔너리만 False
```

```python
    # 논리 연산자
    ## and 연산자 -> 무조건 모든 식이 True 여야 True 출력
print(True and True)
print(True and False)
print(False and True)
print(False and False)
print(True and True and False)
```

```python
    ## or 연산자
print(True or True)
print(True or False)
print(False or True)
print(False or False)
print(True or True or False)
```

```python
    ## not 연산자
print(not True)
```
