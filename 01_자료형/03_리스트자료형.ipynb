```python
    # 리스트는 어떻게 만들고 사용할까?

# 리스트명 = [요소1, 요소2, ...]

a = []
b = [1, 2, 3]
c = ["life", "is", "too", "short"]
d = [1, 2, "life", "is"]
e = [1, 2, ["life", "is"]]

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
```

```python
    # 리스트의 인덱싱과 슬라이싱

a = [1, 2, 3]
print(type(a[2]))
```

```python
b = ['1', '2', '3', 'a', 'b', 'c']
print(b[-1])
print(b[0] + b[2] + b[-1])
```

```python
c = [1, 2, ['a', 'b', ["life", "is"]]]
c[-1][2][0]
```

```python
    # 리스트의 슬라이싱
a= [1, 2, 3, 4, 5]

print(a[0:4])
print(a[0:5])
print(a[0:2])
print(a[0:0])
```

```python
print(a[2:])
print(a[:3])
```

```python
a = [1, 2, 3, ['a', 'b', 'c'], 4, 5]

print(a[3][1:])
```

```python
    # 리스트의 더하기
a = [1, 2, 3]
b = [4, 5, 6]

a + b
```

```python
    # 리스트의 곱하기
a = [1, 2, 3]
b = 3

a * b
```

```python
    # 리스트의 수정과 삭제

a = [1, 2, 3]
a[2] = 4
print(a)
print(a[2])
```

```python
    # del로 리스트의 요소 삭제하기
a = [1, 2, 3]
del a[1]
print(a)
```

```python
a = [1, 2, 3, 4, 5]
del a[2:]
print(a)
```

```python
    # 리스트 관련 함수

    ## 리스트 요소 추가하기 - append
a = [1, 2, 3]
a.append(4)
print(a)
```

```python
a.append([5, 6])
print(a)
```

```python
    # 리스트 정렬 - sort

a = [1, 4, 2, 3]
a.sort()
print(a)
a.sort(reverse=True)
print(a)
```

```python
a = [1, 4, 2, 3]
print(sorted(a))
print(sorted(a, reverse=True))
```

```python
    # 리스트 뒤집기 - reverse

a = ['a', 'c', 'b']

a.reverse()
print(a)
```

```python
    # 인덱스의 반환 - index

a = ['a', 'c', 'b']
a.index("c")
```

```python
    # 리스트의 요소 삽입 - insert

a = [1, 2, 3]
a.insert(0, 4)
print(a)
```

```python
    # 리스트의 요소 제거 - remove

a = [1, 2, 3, 1, 2, 3]
a.remove(3)
a.remove(3)
print(a)
```

```python
    # 리스트의 요소 꺼내기 - pop
a = [1, 2, 3]
result = a.pop()
print(result)
print(a)
```

```python
    # 리스트에 포함된 요소 x의 개수 세기 - count
a = [1, 2, 3, 1]
print(a.count(1))
```

```python
    # 리스트 확장 - extend
a = [1, 2, 3]
a.extend([4, 5]) #append와의 차이점.. -> 리스트의 더하기와 같음
print(a)
```
