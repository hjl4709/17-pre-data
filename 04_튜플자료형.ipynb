```python
    # 튜플은 어떻게 만들까? (tuple)
t1 = () # 튜플의 특징 (1) - 괄호
t2 = (1, ) # 튜플의 특징 (2) - 수정이 불가능, 하나의 요소를 적을 때 (,)를 적어야 함.
t3 = (1, 2, 3)
t4 = 1, 2, 3 # 튜플의 특징 (3) - 괄호 없이도 가능
t5 = ('a', 'b', ('ab', 'cd'))

print(type(t1))
print(type(t2))
print(type(t3))
print(type(t4))
print(type(t5))

print(t1)
print(t2)
print(t3)
print(t4)
print(t5)
```

```python
a = 1, # (,)가 있기 때문에 튜플로 판단됨 / 만약 없다면? int 정수형이 됨
print(type(a))
```

```python
    # 튜플의 요소 값을 지우거나 변경하려고 하면 어떻게 될까?

t1 = 1, 2, 'a', 'b'
del t1[0] # 오류 -> 지우는 것이 불가능
print(t1)
```

```python
    # 튜플의 요소 값을 변경하려고 할 때
t1 = 1, 2, 'a', 'b'
t1[0] = 'c' # 오류 -> 변경하는 것이 불가능
```

```python
    # 튜플 인덱싱
t1 = 1, 2, 'a', 'b'
print(t1[2]) # 리스트의 인덱싱과 동일함
```

```python
    # 튜플 슬라이싱
t1 = 1, 2, 'a', 'b'
print(t1[1:])
```

```python
    # 튜플 더하기, 곱셈
t1 = 1, 2, 'a', 'b'
t2 = 3, 4

print(t1 + t2)
print(t1 * 2)
```

```python
    # 튜플 길이 구하기
print(len(t1))
print(len(t2))
```
